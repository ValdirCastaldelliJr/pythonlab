```python
import pymupdf
from deep_translator import GoogleTranslator

sv_translator = GoogleTranslator(source='sv', target='en')
en_translator = GoogleTranslator(source='en', target='pt')
```

## Translate by Text Block methods

```python
def process_translation(translator: GoogleTranslator, doc):
    ocg_xref = doc.add_ocg("translated", on=True)
    print("Paginas total", len(doc), ":")

    idx = 0
    # Original block
    for page in doc:
        idx += 1
        print("page", idx, ":", end=" ")
        
        blocks = page.get_text("blocks", flags=pymupdf.TEXT_DEHYPHENATE)
        translate_text_block(translator, ocg_xref, page, blocks)

    print("Translated completed")
    return doc

def translate_text_block(translator, ocg_xref, page, blocks):
    bdx = 0

    for block in blocks:
        if bdx % 5 == 0:
            print("|", end="", sep="")
        bdx += 1

        bbox = block[:4]  
        original = block[4] 
        translated = translator.translate(original) 

        # page design
        draw_translated_bock(ocg_xref, page, bbox, translated)
            
    print("bdx", bdx)

def draw_translated_bock(ocg_xref, page, bbox, translated):
    page.draw_rect(bbox, color=None, fill=pymupdf.pdfcolor['white'], oc=ocg_xref)
    page.insert_htmlbox(bbox, translated, oc=ocg_xref)
```
